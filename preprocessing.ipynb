```python
def generate_all_misspellings(word):
    word = str(word)
    if len(word) < 2:
        return [word]  
    
    word = list(word)
    possible_misspellings = set()
    
    for indices in itertools.combinations(range(len(word)), 2):
        word_copy = word[:]
        
        # Swap
        swapped = word_copy[:]
        swapped[indices[0]], swapped[indices[1]] = swapped[indices[1]], swapped[indices[0]]
        possible_misspellings.add("".join(swapped))
        
        # Replace
        for i in indices:
            for char in "abcdefghijklmnopqrstuvwxyz":
                replaced = word_copy[:]
                replaced[i] = char
                possible_misspellings.add("".join(replaced))
        
        # Delete
        deleted = word_copy[:]
        for i in sorted(indices, reverse=True):
            del deleted[i]
        possible_misspellings.add("".join(deleted))
        
        # Insert
        for i in indices:
            for char in "abcdefghijklmnopqrstuvwxyz":
                inserted = word_copy[:]
                inserted.insert(i, char)
                possible_misspellings.add("".join(inserted))
    
    return list(possible_misspellings)
```

```python
data = [[word, misspelling, 0] for word in words for misspelling in generate_all_misspellings(word)]
# half_size = len(data) // 2
# data = random.sample(data, half_size)
```

```python
def generate_totally_incorrect_pairs(words, num_pairs_min=180, num_pairs_max=220):
    different_words = []
    for idx, word in enumerate(words):
        other_words = words[:idx] + words[idx+1:]  
        num_pairs = random.randint(num_pairs_min, num_pairs_max)
        random_250_words = random.sample(other_words, num_pairs)
        for other_word in random_250_words:
            different_words.append([word, other_word, 1])
    return different_words
```

```python
totally_different_words = generate_totally_incorrect_pairs(words)
```

```python
combined_data = totally_different_words + data
random.shuffle(combined_data)
combined_data
```
